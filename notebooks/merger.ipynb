```python
import pandas as pd
import os
```

```python
drought_index_dir = os.listdir('../data/Drought Index')
drought_index_files = [f for f in drought_index_dir if f.startswith('Drought Index') and f.endswith('.csv')]
```

```python
drought_index_files[:5]
```

```python
cimis_dir = os.listdir('../data/CIMIS')
cimis_files = [f for f in cimis_dir if f.endswith('County.csv')]
```

```python
cimis_files[:5]
```

```python
d_list_df = []

for file_name in drought_index_files:
    file_path = os.path.join('../data/Drought Index', file_name)
    df = pd.read_csv(file_path)
    d_list_df.append(df)

drought_index_merged_df = pd.concat(d_list_df, ignore_index=True)
```

```python
drought_index_merged_df
```

```python
c_list_df = []

for file_name in cimis_files:
    file_path = os.path.join('../data/CIMIS', file_name)
    df = pd.read_csv(file_path)
    c_list_df.append(df)

cimis_files_merged_df = pd.concat(c_list_df, ignore_index=True)
```

```python
cimis_files_merged_df.to_csv('../data/cimis_merged.csv', index=False) 
```
